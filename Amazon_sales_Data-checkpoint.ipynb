```python
# ---------------------------------------------------------
# AMAZON SALES REPORT ANALYSIS - COMPLETE PROJECT
# ---------------------------------------------------------

# 📦 Step 1: Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Set visualization style
plt.style.use('seaborn-v0_8-whitegrid')

# ---------------------------------------------------------
# 📥 Step 2: Load Dataset
# ---------------------------------------------------------
# Replace with your dataset path or URL
url = "https://drive.google.com/uc?id=1YrjYKtS1WHmINL6eafRsrDzrZaw2_WvX"  
df = pd.read_csv(url)

# Display basic info
print("✅ Dataset Loaded Successfully!")
print("Shape of data:", df.shape)
print("\nData Preview:\n", df.head())

# ---------------------------------------------------------
# 🧹 Step 3: Data Cleaning (Fixed)
# ---------------------------------------------------------

# Standardize column names
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

# Show available columns
print("\nAvailable Columns:\n", list(df.columns))

# Handle missing values safely
for col in ['amount', 'quantity']:
    if col in df.columns:
        df = df.dropna(subset=[col])
    else:
        print(f"⚠️ Column '{col}' not found, skipping missing value handling for it.")

# Convert date column if present
date_cols = [c for c in df.columns if 'date' in c]
if date_cols:
    df[date_cols[0]] = pd.to_datetime(df[date_cols[0]], errors='coerce')
    print(f"✅ Date column converted: {date_cols[0]}")
else:
    print("⚠️ No date column found!")

# Remove duplicates
df.drop_duplicates(inplace=True)

print("\n✅ Data Cleaning Completed Successfully!")


# ---------------------------------------------------------
# 📊 Step 4: Sales Overview
# ---------------------------------------------------------
print("\n--- Sales Overview ---")
total_sales = df['amount'].sum()
total_orders = df['order_id'].nunique()
avg_order_value = df['amount'].mean()

print(f"Total Sales Amount: ${total_sales:,.2f}")
print(f"Total Orders: {total_orders}")
print(f"Average Order Value: ${avg_order_value:,.2f}")

# Sales Trend Over Time
if 'date' in df.columns:
    sales_trend = df.groupby('date')['amount'].sum().reset_index()
    plt.figure(figsize=(10,5))
    plt.plot(sales_trend['date'], sales_trend['amount'], color='blue')
    plt.title('Sales Trend Over Time')
    plt.xlabel('Date')
    plt.ylabel('Sales Amount')
    plt.show()

# ---------------------------------------------------------
# 🛍️ Step 5: Product Analysis
# ---------------------------------------------------------
print("\n--- Product Analysis ---")

if 'product_category' in df.columns:
    top_products = df['product_category'].value_counts().head(10)
    print("\nTop Product Categories:\n", top_products)

    plt.figure(figsize=(8,5))
    sns.barplot(x=top_products.values, y=top_products.index, palette='viridis')
    plt.title('Top 10 Product Categories')
    plt.xlabel('Number of Orders')
    plt.ylabel('Product Category')
    plt.show()

# Size Distribution
if 'size' in df.columns:
    plt.figure(figsize=(6,4))
    df['size'].value_counts().head(10).plot(kind='bar', color='orange')
    plt.title('Product Size Distribution')
    plt.xlabel('Size')
    plt.ylabel('Count')
    plt.show()

# ---------------------------------------------------------
# 🚚 Step 6: Fulfillment Analysis
# ---------------------------------------------------------
print("\n--- Fulfillment Analysis ---")

if 'fulfilment' in df.columns:
    fulfillment_perf = df['fulfilment'].value_counts()
    print(fulfillment_perf)

    plt.figure(figsize=(6,4))
    fulfillment_perf.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=sns.color_palette('Set2'))
    plt.title('Fulfillment Method Distribution')
    plt.ylabel('')
    plt.show()

# ---------------------------------------------------------
# 👥 Step 7: Customer Segmentation
# ---------------------------------------------------------
print("\n--- Customer Segmentation ---")

if 'ship-city' in df.columns:
    city_sales = df.groupby('ship-city')['amount'].sum().sort_values(ascending=False).head(10)
    print("\nTop 10 Cities by Sales:\n", city_sales)

    plt.figure(figsize=(8,5))
    sns.barplot(x=city_sales.values, y=city_sales.index, palette='coolwarm')
    plt.title('Top 10 Cities by Sales Amount')
    plt.xlabel('Sales Amount')
    plt.ylabel('City')
    plt.show()

# ---------------------------------------------------------
# 🗺️ Step 8: Geographical Analysis
# ---------------------------------------------------------
print("\n--- Geographical Analysis ---")

if 'ship-state' in df.columns:
    state_sales = df.groupby('ship-state')['amount'].sum().sort_values(ascending=False).head(10)
    print("\nTop 10 States by Sales:\n", state_sales)

    plt.figure(figsize=(8,5))
    sns.barplot(x=state_sales.values, y=state_sales.index, palette='crest')
    plt.title('Top 10 States by Sales Amount')
    plt.xlabel('Sales Amount')
    plt.ylabel('State')
    plt.show()

# ---------------------------------------------------------
# 💡 Step 9: Business Insights
# ---------------------------------------------------------
print("\n--- Business Insights & Recommendations ---")
print("""
1️⃣ High-performing product categories should be prioritized for promotions and ads.
2️⃣ States and cities with lower sales may need localized marketing campaigns.
3️⃣ Analyze low-performing fulfillment methods to improve delivery efficiency.
4️⃣ Consider bundling top-selling products to boost average order value.
5️⃣ Maintain adequate inventory for top 10 products and popular sizes.
6️⃣ Track daily or monthly sales trends to anticipate seasonal peaks.
""")

# ---------------------------------------------------------
# ✅ End of Project
# ---------------------------------------------------------
print("\n🎯 Amazon Sales Analysis Completed Successfully!")

```
